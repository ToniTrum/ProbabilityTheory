```python
import numpy as np
```

```python
n = 30
```

```python
inputs = [
    6,
    3.3333,
    0.3333,
    0.15,
    1.2609,
    0.875,
    0.5789,
    2,
    2.1429,
    0.8182,
    1.1818,
    3.75,
    0.5556,
    0.6923,
    2.5,
    2.1667,
    1.4286,
    0.5,
    8,
    2.3333,
    1.6667,
    0.4,
    2.6667,
    1.6667,
    1.2,
    0.5,
    2,
    0.8571,
    0.9333,
    6
]
```

```python
x_mean = sum(inputs) / n
x_mean
```

```python
dispersion = sum((x - x_mean) ** 2 for x in inputs) / n
dispersion
```

Общая дисперсия $D = 3.34$ характеризует меру рассеивания $x_i$ относительно общей средней $\overline{x} = 1.95$. 
Чем дисперсия больше, тем дальше разбросаны $x_i$ от средней $\overline{x}$

```python
k = int(1 + 3.322 * np.log10(n))
k
```

```python
h = (max(inputs) - min(inputs)) / k
h
```

```python
inputs = sorted(inputs)
inputs
```

```python
groups = [inputs[i: i + k + 1] for i in range(0, len(inputs), k + 1)]
groups
```

```python
groups[0].append(groups[1].pop(0))
groups[-2].append(groups[-1].pop(0))
groups
```

```python
n_i = [len(group) for group in groups]
n_i
```

```python
x_mean_i = [sum(group) / len(group) for group in groups]
x_mean_i
```

```python
squared_deviation = [sum((x - x_mean_i[i]) ** 2 for x in group) for i, group in enumerate(groups)]
squared_deviation
```

```python
dispersion_i = [squared_dev / n_i[i] for i, squared_dev in enumerate(squared_deviation)]
dispersion_i
```

Групповая дисперсия характеризует меру разброса значений группы, относительно групповой средней. Наименьшая дисперсия получилась по 2-й группе, это означает, что варианты $x_i$ этой группы расположены достаточно близко к средне групповой. Максимальная дисперсия в 5 группе, это означает, что варианты $x_i$ этой группы расположены далеко от средней групповой.

```python
ingroup_dispersion = sum((dispersion_i[i]) * n_i[i] for i in range(len(n_i))) / n
ingroup_dispersion
```

Внутригрупповая дисперсия характеризует средневзвшанную вариацию значений $x_i$ по группам.

```python
a = [(x_mean - x_mean_i[i]) ** 2 for i in range(len(x_mean_i))]
a
```

```python
squared_deviation_n_i = [squared_dev * n_i[i] for i, squared_dev in enumerate(a)]
squared_deviation_n_i
```

```python
aroundgroup_dispersion = sum(squared_deviation_n_i) / n
aroundgroup_dispersion
```

Межгрупповая дисперсия характеризует меру разброса групповых средних относительно общих средних. Чем эта дисперсия больше, тем дальше расположены групповые средние от общей средней.

Этот коэффициент харктеризует долю вариации, объяснённую группировочным фактором. Коэффициент детерминации

Чем больше межгрупповая дисперсия, тем больше коэффициент детерминации

85 % металлоёмкости станков объясняются тем фактором, что станки принадлежат разным производителям, оставшиеся 15% вариации объясняются другими причинами
