# Импорты

```python
import numpy as np
from scipy import stats
import pandas as pd
```

# 1. Выбор варианта

```python
random_state = 18
print(random_state % 6)
```

# 2. Оценка параметра распределения Бернулли методом моментов

$\displaystyle
X \sim Bernoulli(p): X_1, X_2, \ldots, X_n \\
F_X \in \{F_X(x, p) | p \in [0; 1] \} \\
E(X) = p = \overline{x}
$

Оценка параметра $\^{p}_n = \overline{x}$

# 3. Оценка параметра распределения Бернулли методом максимального правдоподобия

$\displaystyle
X \sim Bernoulli(p): X_1, X_2, \ldots, X_n \\
F_X \in \{F_X(x, p) | p \in [0; 1] \} \\
p_X(x) = p^x (1 - p)^{1 - x} \\
L(p, x_1, \ldots, x_n) = \prod_{i = 1}^n p^{x_i} (1 - p)^{1 - x_i} = p^{\scriptsize \displaystyle \sum_{i = 1}^n x_i} (1 - p)^{\scriptsize \displaystyle n - \sum_{i = 1}^n x_i} \\
\ln L(p, x_1, \ldots, x_n) = \ln \left( p^{\scriptsize \displaystyle \sum_{i = 1}^n x_i} (1 - p)^{\scriptsize \displaystyle n - \sum_{i = 1}^n x_i} \right) = \sum_{i = 1}^n x_i \ln p + \left( n - \sum_{i = 1}^n x_i \right) \ln(1 - p) \\
\^{p}_n = \argmax_{p \in [0; 1]} \ln L(p, x_1, \ldots, x_n)
$

$\displaystyle
\frac{d\ln L}{dp} = \frac{\displaystyle \sum_{i = 1}^n x_i}{p} + \frac{\displaystyle \sum_{i = 1}^n x_i - n}{1 - p} = 0 \\
(1 - p) \sum_{i = 1}^n x_i + \left( \sum_{i = 1}^n x_i - n \right) p = 0 \\
pn = \sum_{i = 1}^n x_i \\
p = \sum_{i = 1}^n \frac{x_i}{n} \\
\frac{d^2 \ln L}{dp^2} = \underbrace{-\frac{\displaystyle \sum_{i = 1}^n x_i}{p^2}}_{\displaystyle \leq 0} + \underbrace{\frac{\displaystyle \sum_{i = 1}^n x_i - n}{(1 - p)^2}}_{\displaystyle \leq 0} \leq 0
$

$\displaystyle
\text{Оценка параметра } \^{p}_n = \sum_{i = 1}^n \frac{x_i}{n} = \overline{x}
$

# 4. Генерация выборок

```python
data_100 = stats.bernoulli.rvs(p=0.25, size=100, random_state=random_state)
data_10_000 = stats.bernoulli.rvs(p=0.25, size=10_000, random_state=random_state)
```

# 5. Табличный результат

```python
df = pd.DataFrame({
    "Размер выборки": [
        100, 10_000
    ],
    "Истинный параметр": [
        0.25, 0.25
    ],
    "Оценка метода моментов": [
        np.mean(data_100),
        np.mean(data_10_000)
    ],
    "Оценка метода максимального момента": [
        np.mean(data_100),
        np.mean(data_10_000)
    ]
})
df
```
