# Импорты

```python
import numpy as np
import scipy.stats as scs
```

# Таблица

```python
X = np.array([45.1, 59.0, 57.2, 61.8, 58.8, 47.2, 55.2])
Y = np.array([68.8, 61.2, 59.9, 56.7, 55.0, 54.3, 49.3])
```

# Регрессии

## 1. Линейная регрессия

$\displaystyle
y = a + bx \\
a = \overline{Y} - b \overline{X} \\
b = \frac{\overline{YX} - \overline{Y} \overline{X}}{\sigma_X^2}
$

```python
n = len(X)
X_mean = np.mean(X)
X_var = np.var(X)
Y_mean = np.mean(Y)
Y_var = np.var(Y)
YX_mean = np.mean(X * Y)

b = (YX_mean - Y_mean * X_mean) / X_var
a = Y_mean - b * X_mean
y_pred = a + b * X
print(f"y = {a:.2f} {"+" if b >= 0 else "-"} {np.abs(b):.2f}x")

r_XY = np.corrcoef(X, Y)[0, 1]
print(f"r_XY = {r_XY:.2f}")

A = 1 / n * np.sum(np.abs((Y - y_pred) / Y)) * 100
print(f"A = {A:.2f} %")

F_comp = r_XY ** 2 / (1 - r_XY ** 2) * (n - 2)
F_table = scs.f.ppf(0.95, dfn=1, dfd=n - 2)
print(f"F_расч = {F_comp:.2f}")
print(f"F_табл = {F_table:.2f}")
print(f"H_0 {"не " if F_comp <= F_table else ""}отвергаем")
```

## 2. Степенная регрессия

$\displaystyle
y = a x^b \\
\lg y = \lg a + b \lg x \\
\~{y} = \~{a} + b \~{x} \\
a = 10^{\~{a}} \\
\~{a} = \overline{\~{Y}} - b \overline{\~{X}} \\
b = \frac{\overline{\~{Y}\~{X}} - \overline{\~{Y}} \overline{\~{X}}}{\sigma_{\~{X}}^2}
$

```python
n = len(X)
lg_X = np.log10(X)
lg_Y = np.log10(Y)
lg_X_mean = np.mean(lg_X)
lg_X_var = np.var(lg_X)
lg_Y_mean = np.mean(lg_Y)
lg_Y_var = np.var(lg_Y)
lg_Y_lg_X_mean = np.mean(lg_X * lg_Y)

b = (lg_Y_lg_X_mean - lg_Y_mean * lg_X_mean) / lg_X_var
lg_a = lg_Y_mean - b * lg_X_mean
a = 10 ** lg_a
lg_y_pow = lg_a + b * lg_X
y_pred = 10 ** lg_y_pow
print(f"y = {a:.2f}x^{b:.2}")

r_lg_X_lg_Y = np.corrcoef(lg_X, lg_Y)[0, 1]
r_XY = r_lg_X_lg_Y
print(f"r_XY = {r_XY:.2f}")

A = 1 / n * np.sum(np.abs((Y - y_pred) / Y)) * 100
print(f"A = {A:.2f} %")

F_comp = r_XY ** 2 / (1 - r_XY ** 2) * (n - 2)
F_table = scs.f.ppf(0.95, dfn=1, dfd=n - 2)
print(f"F_расч = {F_comp:.2f}")
print(f"F_табл = {F_table:.2f}")
print(f"H_0 {"не " if F_comp <= F_table else ""}отвергаем")
```

## 3. Показательная регрессия

$\displaystyle
y = a b^x \\
\lg y = \lg a + x \lg b \\
\~{y} = \~{a} + \~{b} x \\
a = 10^{\~{a}} \\
b = 10^{\~{b}} \\
\~{a} = \overline{\~{Y}} - \~{b} \overline{X} \\
\~{b} = \frac{\overline{\~{Y}X} - \overline{\~{Y}} \overline{X}}{\sigma_{X}^2}
$

```python
n = len(X)
lg_Y = np.log10(Y)
X_mean = np.mean(X)
X_var = np.var(X)
lg_Y_mean = np.mean(lg_Y)
lg_Y_var = np.var(lg_Y)
lg_Y_X_mean = np.mean(X * lg_Y)

lg_b = (lg_Y_X_mean - lg_Y_mean * X_mean) / X_var
lg_a = lg_Y_mean - lg_b * X_mean
a = 10 ** lg_a
b = 10 ** lg_b
lg_y_exp = lg_a + lg_b * X
y_pred = 10 ** lg_y_exp
print(f"y = {a:.2f} * {b:.2}^x")

r_X_lg_Y = np.corrcoef(X, lg_Y)[0, 1]
r_XY = r_X_lg_Y
print(f"r_XY = {r_XY:.2f}")

A = 1 / n * np.sum(np.abs((Y - y_pred) / Y)) * 100
print(f"A = {A:.2f} %")

F_comp = r_XY ** 2 / (1 - r_XY ** 2) * (n - 2)
F_table = scs.f.ppf(0.95, dfn=1, dfd=n - 2)
print(f"F_расч = {F_comp:.2f}")
print(f"F_табл = {F_table:.2f}")
print(f"H_0 {"не " if F_comp <= F_table else ""}отвергаем")
```

## 4. Равномерная гипербола

$\displaystyle
y = a + \frac{b}{x} \\
y = a + b x^{-1} \\
y = a + b \~{x} \\
a = \overline{Y} - b \overline{\~{X}} \\
b = \frac{\overline{Y\~{X}} - \overline{Y} \overline{\~{X}}}{\sigma_{\~{X}}^2}
$

```python
n = len(X)
X_1 = 1 / X
X_1_mean = np.mean(X_1)
X_1_var = np.var(X_1)
Y_mean = np.mean(Y)
Y_var = np.var(Y)
YX_1_mean = np.mean(X_1 * Y)

b = (YX_1_mean - Y_mean * X_1_mean) / X_1_var
a = Y_mean - b * X_1_mean
y_pred = a + b * X_1
print(f"y = {a:.2f} {"+" if b >= 0 else "-"} {np.abs(b):.2f}/x")

r_X_1Y = np.corrcoef(X_1, Y)[0, 1]
r_XY = r_X_1Y
print(f"r_XY = {r_XY:.2f}")

A = 1 / n * np.sum(np.abs((Y - y_pred) / Y)) * 100
print(f"A = {A:.2f} %")

F_comp = r_XY ** 2 / (1 - r_XY ** 2) * (n - 2)
F_table = scs.f.ppf(0.95, dfn=1, dfd=n - 2)
print(f"F_расч = {F_comp:.2f}")
print(f"F_табл = {F_table:.2f}")
print(f"H_0 {"не " if F_comp <= F_table else ""}отвергаем")
```
