# Импорты

```python
import numpy as np
import scipy.stats as scs 
```

# Входные данные

```python
X = np.array([
    19.7,
    18.7,
    16.3,
    18.7,
    19.3,
    18.8,
    19.3,
    18.8,

    16.3,
    17.4,
    21.8,
    18.8,
    20.2,
    19.3,
    18.4,
    19.3,

    21.8,
    18.1,
    19.4,
    19.7,
    21.8,
    18.8,
    18.7,
    16.3
])
```

# Характеристики положения

```python
print("Среднее:", np.mean(X))
print("Медиана:", np.median(X))
print("Мода:", scs.mode(X).mode)
```

# Характеристики рассеяния

```python
print("Наибольшее:", max(X))
print("Наименьшее:", min(X))
print("Размах:", max(X) - min(X))
print("Выборочная дисперсия:", np.var(X, ddof=1))
print("Генеральная дисперсия:", np.var(X))
print("Выборочное среднеквадратичное отклонение:", np.std(X, ddof=1))
print("Генеральное среднеквадратичное отклонение:", np.std(X))
```

# Характеристики формы

```python
print("Эксцесс:", scs.kurtosis(X))
print("Коэффициент асимметрии:", scs.skew(X))
```
