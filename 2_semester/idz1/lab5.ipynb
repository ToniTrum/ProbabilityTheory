# Импорты

```python
import numpy as np
import scipy.stats as scs
```

# Входные данные

```python
X = np.array([
    19.3,
    19.9,
    23.1,
    18.8,
    17.4,
    21.6,
    19.1,
    22.5,
    20.2,

    18.3,
    16.4,
    17.3,
    18.3,
    15.8,
    21.2,
    19.3,
    21.6,
    19.9,

    20.6,
    19.4,
    18.7,
    16.3,
    18.4,
    19.3,
    18.8,
    17.8,
    20.5
])
```

# Статистика

```python
print("Среднее:", np.mean(X))
print("Статистическая ошибка:", scs.sem(X))
print("Медиана:", np.median(X))
print("Мода:", scs.mode(X))
print("Стандартное отклонение:", np.std(X))
print("Дисперсия:", np.var(X))
print("Эксцесс:", scs.kurtosis(X))
print("Ассиметричность:", scs.skew(X))
print("Интервал:", max(X) - min(X))
print("Минимум:", min(X))
print("Максимум:", max(X))
print("Сумма:", np.sum(X))
print("Счёт:", len(X))
```
