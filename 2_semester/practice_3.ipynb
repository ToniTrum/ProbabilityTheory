```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "1": [200, 140, 170, 145, 165],
    "2": [190, 150, 210, 150, 150],
    "3": [230, 190, 200, 190, 200],
    "4": [150, 170, 150, 170, 180]
})
df
```

```python
df_mean = df.mean()
df_mean
```

```python
df_var = df.var()
df_var
```

```python
x_mean = df_mean.mean()
x_mean
```

```python
D_megh = 5 * sum((df_mean - x_mean) ** 2)
D_megh
```

```python
D_vn = sum(df_var * 5) / 20
D_vn
```
